# FFT Basic Demo (Rust / evcxr)

This notebook demonstrates basic FFT concepts using Rust.
These are clean demos; for full exploration see `research/notebooks/`. 

````python
// Import crates in evcxr
:dep ndarray = "0.15"
:dep rustfft = "6"
use rustfft::{FftPlanner, num_complex::Complex};

fn make_signal(n: usize) -> Vec<Complex<f32>> {
    let sr = n as f32;
    (0..n).map(|i| {
        let t = i as f32 / sr;
        let s = (2.0*std::f32::consts::PI*5.0*t).sin() + 0.5*(2.0*std::f32::consts::PI*10.0*t).sin();
        Complex::new(s, 0.0)
    }).collect()
}

let mut x = make_signal(1024);
println!("len={} first={:?}", x.len(), x[0]);
````

````python
// Compute FFT using rustfft
let mut planner = FftPlanner::<f32>::new();
let fft = planner.plan_fft_forward(x.len());
fft.process(&mut x);

// Magnitude spectrum (first 16 bins)
let mags: Vec<f32> = x.iter().map(|c| c.norm()).take(16).collect();
mags
````

````python
// Inverse FFT sanity check
let n = x.len();
let mut planner = FftPlanner::<f32>::new();
let ifft = planner.plan_fft_inverse(n);
ifft.process(&mut x);
// Normalize by n after inverse
for v in &mut x { *v /= n as f32; }
// Show first 5 reconstructed samples
x.iter().take(5).map(|c| c.re).collect::<Vec<f32>>()
````
