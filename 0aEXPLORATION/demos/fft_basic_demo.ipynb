# FFT Basic Demo

This notebook demonstrates basic FFT operations using Candle tensors.
This is a clean demo suitable for upstream review - see `research/notebooks/` for full exploration.

````python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt

# This would use candle-pyo3 when available
# For now, demonstrate the concept with numpy
````

````python
# Create a simple signal
t = np.linspace(0, 1, 1000)
signal = np.sin(2 * np.pi * 5 * t) + 0.5 * np.sin(2 * np.pi * 10 * t)

print(f"Signal shape: {signal.shape}")
print(f"Sample rate: {len(signal)} Hz")
````

````python
# Compute FFT
fft_result = np.fft.fft(signal)
frequencies = np.fft.fftfreq(len(signal), 1/len(signal))

print(f"FFT result shape: {fft_result.shape}")
print(f"Frequency bins: {len(frequencies)}")
````
