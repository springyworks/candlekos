# Candle + evcxr quickstart (local repo)

This notebook demonstrates using Candle directly from your local repository, via the Rust evcxr Jupyter kernel.

- Kernel: select the Rust (evcxr) kernel.
- Dependencies: we point evcxr to the local `candle-core` and `candle-nn` crates using `:dep` directives with absolute paths.
- Execution: all cells are CPU by default. CUDA usage is optional and noted below.

If you don't have the Rust Jupyter kernel installed yet, install `evcxr_jupyter` and restart Jupyter before running.

```rust
// evcxr setup: point to local Candle crates via absolute paths.
// Adjust the root path if you moved the repo.
:dep anyhow = "1"
:dep thiserror = "1"

// Absolute path to the repo root
let repo = std::path::PathBuf::from("/home/rustuser/projects/rust/from_github/candle");
assert!(repo.exists(), "Repo path not found: {:?}", repo);

// Build paths to core and nn crates
let core = repo.join("candle-core");
let nn = repo.join("candle-nn");
assert!(core.exists() && nn.exists(), "Missing local candle crates: core={:?} nn={:?}", core, nn);

// Tell evcxr to depend on local crates
:dep candle-core = { path = "/home/rustuser/projects/rust/from_github/candle/candle-core" }
:dep candle-nn = { path = "/home/rustuser/projects/rust/from_github/candle/candle-nn" }

// Optional: if the above path deps fail (e.g., running on another machine), you can fall back to crates.io
// by evaluating the lines below instead of the :dep path ones (comment out one set or the other):
// :dep candle-core = "0.5"
// :dep candle-nn = "0.5"

use candle_core as candle;
use candle_nn as nn;
```

```rust
use candle::{Device, Tensor, DType, Result};

// Small CPU tensor, elementwise sin
a = Tensor::arange(0f32, 16f32, &Device::Cpu)?.reshape((4, 4))?;
let s = a.sin()?;
println!("a shape: {:?}", a.dims());
println!("s shape: {:?}", s.dims());
println!("a: {:?}", a.to_vec2::<f32>()?);
println!("sin(a): {:?}", s.to_vec2::<f32>()?);
```

```rust
// Device and dtype summary
let dev = Device::Cpu;
println!("device: {:?}", dev);
println!("default dtype: {:?}", DType::F32);
let ones = Tensor::ones((2, 2), DType::F32, &dev)?;
println!("ones: {:?}", ones.to_vec2::<f32>()?);
```
