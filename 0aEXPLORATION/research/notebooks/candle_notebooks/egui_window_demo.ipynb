# egui_window Demo

Showcases the lightweight HTML 'egui_window' mock utilities. These do not open a native GUI; they render styled HTML blocks inside the notebook output.

```rust
:toolchain stable
:dep candle-notebooks = { version = "0.1.0", features = ["external-window"] }
// Unified setup: single dependency, imports, and initialization.
use candle_notebooks as nb;
use nb::{Device, Tensor};
use nb::egui_window;
use std::fs;
let device = Device::Cpu;
nb::set_notebook_cwd().expect("failed to set notebook cwd");
nb::set_image_store_rel_dir("images_store").unwrap();
fs::create_dir_all("images_store").ok();
println!("Toolchain: stable | Device: {:?}", device);
println!("CWD: {}", std::env::current_dir().unwrap().display());
println!("Image store: images_store/");
println!("candle-notebooks features: external-window");
```

```rust
// Basic text window.
egui_window::open_text("Hello Window", "This is a plain text styled block.");
// Color squares window.
egui_window::open_color_squares("Palette");
```

```rust
// Generate a tensor gradient and embed as inline image in a window.
let h: usize = 96; let w: usize = 160;
let mut r: Vec<f32> = Vec::with_capacity(h*w);
let mut g: Vec<f32> = Vec::with_capacity(h*w);
let mut b: Vec<f32> = Vec::with_capacity(h*w);
for y in 0..h { for x in 0..w { let fx = x as f32/(w as f32-1.0); let fy = y as f32/(h as f32-1.0); r.push(fx); g.push(fy); b.push((fx+fy)/2.0); }}
let dev: Device = Device::Cpu;
let r: Tensor = Tensor::from_vec(r, (1_usize,h,w), &dev).unwrap();
let g: Tensor = Tensor::from_vec(g, (1_usize,h,w), &dev).unwrap();
let b: Tensor = Tensor::from_vec(b, (1_usize,h,w), &dev).unwrap();
let rgb: Tensor = Tensor::cat(&[&r, &g, &b], 0).unwrap();
let url: String = nb::tensor_to_png_data_url_rgb(&rgb).unwrap();
let html: String = format!("<p style='margin:4px 0 8px 0;font-weight:500;'>RGB Gradient Tensor</p><img style='image-rendering:pixelated;border:1px solid #333;max-width:100%;' src='{}'/>", url);
// Use external window if available; will fallback to inline if feature is disabled.
egui_window::open_external("Tensor Image", &html);
// close the seperate external egui-window to continue
```

```rust
// Create multiple windows in one cell (OK).
for i in 0..3 {
  let body = format!("Window index {} -- generated at runtime.", i);
  egui_window::open_text(&format!("Batch {}", i), &body);
}
```
